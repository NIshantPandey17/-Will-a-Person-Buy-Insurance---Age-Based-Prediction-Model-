```python
import pandas as pd
```

```python
df= pd.read_csv("insurance_data.csv")
```

```python
df.head(2)
```

```python
x=df[['age']]
y=df.bought_insurance
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("insurance brought analyzer")
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x,y)
```

```python
model.predict([[20]])
```

```python
import pickle
pickle.dump(model,open("insurance_brought."))
```
